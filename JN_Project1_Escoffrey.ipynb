```python
# Import dependencies
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read in veteran suicide by sex file
vet_suicide_sex = "../Project_1_Team_2/Veteran Suicides by Sex.csv"
vet_suicide_df = pd.read_csv(vet_suicide_sex)
vet_suicide_df.head()
```

```python
#Sum of veteran female & male suicides by region and year
vet_groupby = vet_suicide_df.groupby(["Region","Year"])
vet_groupby_sum = vet_groupby.sum()
vet_groupby_sum.reset_index()
```

```python
# read in suicide by age file
vet_suicide_age = "../Project_1_Team_2/Suicides by Age.csv"
vet_age_df = pd.read_csv(vet_suicide_age)
vet_age_df.head()
```

```python
# dropping state rows
#Veteran and General Population Suicides by age
vet_age_df2 = vet_age_df.loc[vet_age_df["State of Death"].isin(["Total U.S."])]
vet_age_df2.head(20)
```

```python
# read in veteran suicide by method file
vet_method = "../Project_1_Team_2/Suicides by Method.csv"
vet_method_df = pd.read_csv(vet_method)
vet_method_df.head()
```

```python
# dropping state rows
vet_method_df2 = vet_method_df.loc[vet_method_df["State of Death"]=="All"]
vet_method_df2.head(20)
```

```python
# read civilian female_suicide_method file and clean the data
female_suicide_method = "../Project_1_Team_2/female_suicide_methods.csv"
female_suicide_method_df = pd.read_csv(female_suicide_method)
female_suicide_method_df = female_suicide_method_df.drop(columns = ['Percent'])

female_suicide_method_df.head(20)
```

```python
# read civilian male_suicide_method file and clean the data
male_suicide_method = "../Project_1_Team_2/male_suicide_methods.csv"
male_suicide_method_df = pd.read_csv(male_suicide_method)
male_suicide_method_df.head()
male_suicide_method_df = male_suicide_method_df.drop(columns=['Last Year', 'Percent','Race/Ethnicty','Age Group'])
male_suicide_method_df = male_suicide_method_df.rename(columns ={'First Year': 'Year'})

male_suicide_method_df.head(20)
```

```python
#Merge male and female civilan method data, clean and rename dataframe
general_population_suicide_method = pd.merge(male_suicide_method_df,female_suicide_method_df, 
                                             on=['Cause of Death', 'Year','State'],
                                             suffixes=('_male', '_female'))

general_population_suicide_method['Deaths'] = general_population_suicide_method['Deaths_male'] + general_population_suicide_method['Deaths_female']
general_population_suicide_method.head()
```

```python
#Dropped columns not needed 
#Civilian data
general_population_suicide_method = general_population_suicide_method.drop(columns = ['Deaths_male','Sex_male','Deaths_female','Sex_female','Age Group'])
general_population_suicide_method.head()
```

```python
general_population_suicide_method['Cause of Death'].value_counts()
```

```python
#Combine all other/unspecified causes of death
general_population_suicide_method['Cause of Death'] = general_population_suicide_method['Cause of Death'].replace({"Other Spec., classifiable": "Unspecified",
                                                                                                                       "Other Spec., NEC":"Unspecified"})
general_population_suicide_method.head(20) 
```

```python
#Create data frame of only unspecified causes of death
unspecified_df=general_population_suicide_method.loc[general_population_suicide_method
    ["Cause of Death"]=="Unspecified",:]

unspecified_df.head()
```

```python
#Calculate sum of unspecified causes by state and year
unspecified_df=unspecified_df.groupby(["State","Year"]).sum()
unspecified_df.head(20)
```

```python
#add cause of death index for unspecified causes
unspecified_df["Cause of Death"]="Unspecified"
unspecified_df.reset_index(inplace=True)
```

```python
#reorder columns
unspecified_df=unspecified_df[["Cause of Death","State","Year","Deaths"]]
unspecified_df.head()
```

```python
#create data frame for specified causes
specified_df=general_population_suicide_method.loc[general_population_suicide_method["Cause of Death"]!="Unspecified",:]
specified_df.head()
```

```python
#append specified and unspecified data frames
#specified_df.append(unspecified_df)
frames = [specified_df,unspecified_df]
specified_df = pd.concat(frames)
```

```python
specified_groupby_df = specified_df.groupby(["State","Year"]).sum()

specified_groupby_df.head()


civ_methods = pd.merge(specified_df,specified_groupby_df, on=['State', 'Year'])
civ_methods=civ_methods.rename(columns={"Deaths_x":"Deaths","Deaths_y":"Total Deaths"})
civ_methods.head()
```

```python
#Calculate Percentage
civ_methods['Percent']=round(civ_methods['Deaths']/civ_methods['Total Deaths']*100,2) 
civ_methods.head(20)
```

```python
# save the general_population_suicide_method data to csv
civ_methods.to_csv('../Project_1_Team_2/general_population_suicide_methods.csv',index=False)
```

```python
# Convert object data type to int64
millennial_df.replace(',', '',regex=True, inplace=True)
millennial_df
```

```python
# Convert object data type to string
millennial_df['Veteran\nSuicides']=millennial_df['Veteran\nSuicides'].astype('int64')
millennial_df['Veteran\nSuicide\nRate\nper\n100,000']=millennial_df['Veteran\nSuicide\nRate\nper\n100,000'].astype('float')
millennial_df['General\nPopulation\nSuicides']=millennial_df['General\nPopulation\nSuicides'].astype('int64')
millennial_df['General\nPopulation\nRate per\n100,000']=millennial_df['General\nPopulation\nRate per\n100,000'].astype('float')
millennial_df.dtypes
```

```python
# Pulling age group 18-34 into new dataframe
millennial_df2 = millennial_df.loc[millennial_df['Age\nGroup']=='18-34']
millennial_df2.head()
```

```python
# List new dataframe
millennial_df2.index=millennial_df2['Year']
list(millennial_df2)
```

```python
# Total Millennial Veteran Suicides vs Total General Population Sucides
ax = millennial_df2[['Veteran\nSuicides','General\nPopulation\nSuicides']].plot(kind='bar')
ax.set(title = 'Veteran Suicide vs. General Population',
       xlabel = 'Year',
       ylabel = 'Number of Suicides')
ax.legend(loc = 'best', frameon=True)

# Save the Figure
plt.savefig("../Project_1_Team_2/Total Millennial Veteran Suicide vs. General Population.png",bbox_inches='tight')
```

```python
# Percentage of Millennial Veteran Suicides vs General Population Sucides per 100,000
ax = millennial_df2[['Veteran\nSuicide\nRate\nper\n100,000','General\nPopulation\nRate per\n100,000']].plot(kind='bar')
ax.set(title = 'Percentage of Veteran Suicides vs. General Population',
       xlabel = 'Year',
       ylabel = 'Percentage of Suicides')
ax.legend(loc = 'best', frameon=True)

# Save the Figure
plt.savefig("../Project_1_Team_2/Percentage Millennial Veteran Suicide vs. General Population.png",bbox_inches='tight')
```
