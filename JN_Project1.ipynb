```python
# read in suicide by age file
vet_suicide_age = "../Project_1_Team_2/Suicides by Age.csv"
vet_age_df = pd.read_csv(vet_suicide_age)
vet_age_df.head()
```

```python
# dropping state rows
vet_age_df2 = vet_age_df.loc[vet_age_df["State of Death"].isin(["Total U.S.","All"])]
vet_age_df2.head(20)
```

```python
# read in suicide by method file
vet_method = "../Project_1_Team_2/Suicides by Method.csv"
vet_method_df = pd.read_csv(vet_method)
vet_method_df.head()
```

```python
# dropping state rows
vet_method_df2 = vet_method_df.loc[vet_method_df["State of Death"]=="All"]
vet_method_df2.head(20)
```

```python
# read female_suicide_method file and clean the data
female_suicide_method = "../Project_1_Team_2/female_suicide_methods.csv"
female_suicide_method_df = pd.read_csv(female_suicide_method)
female_suicide_method_df = female_suicide_method_df.drop(columns = ['Percent'])

female_suicide_method_df.head(20)
```

```python
# read male_suicide_method file and clean the data
male_suicide_method = "../Project_1_Team_2/male_suicide_methods.csv"
male_suicide_method_df = pd.read_csv(male_suicide_method)
male_suicide_method_df.head()
male_suicide_method_df = male_suicide_method_df.drop(columns=['Last Year', 'Percent','Race/Ethnicty','Age Group'])
male_suicide_method_df = male_suicide_method_df.rename(columns ={'First Year': 'Year'})

male_suicide_method_df.head(20)
```

```python
#Merge , clean and rename dataframe
general_population_suicide_method = pd.merge(male_suicide_method_df,female_suicide_method_df, 
                                             on=['Cause of Death', 'Year','State'],
                                             suffixes=('_male', '_female'))

general_population_suicide_method['Deaths'] = general_population_suicide_method['Deaths_male'] + general_population_suicide_method['Deaths_female']
general_population_suicide_method.head()
```

```python
general_population_suicide_method = general_population_suicide_method.drop(columns = ['Deaths_male','Sex_male','Deaths_female','Sex_female','Age Group'])
general_population_suicide_method.head()
```

```python
general_population_suicide_method['Cause of Death'].value_counts()
```

```python
#Calculate percentage
general_population_suicide_method_group = general_population_suicide_method.groupby(["State","Year"]).sum()
general_population_suicide_method_group= general_population_suicide_method_group.rename(columns ={'Deaths': 'Total Deaths'})
general_population_suicide_method_group.reset_index()

general_population_suicide_method_C = pd.merge(general_population_suicide_method,general_population_suicide_method_group, 
                                             on=['State', 'Year'])
        
 
general_population_suicide_method_C['Cause of Death'] = general_population_suicide_method_C['Cause of Death'].replace({"Other Spec., classifiable": "Unspecified",
                                                                                                                       "Other Spec., NEC":"Unspecified"})
general_population_suicide_method_C.head(20) 
```

```python
general_population_suicide_method_C['Percent']=round(general_population_suicide_method_C['Deaths']/general_population_suicide_method_C['Total Deaths']*100,2) 
general_population_suicide_method_C.head()
general_population_suicide_method_C['Cause of Death'].value_counts()
```

```python
general_population_suicide_method_C['Cause of Death'] = general_population_suicide_method_C['Cause of Death'].replace({"Other Spec., classifiable": "Other","Other Spec., NEC":"Other","Unspecified":"Other"})
general_population_suicide_method_C.head(20)
general_population_suicide_method_C
```

```python
# save the general_population_suicide_method data to csv
general_population_suicide_method_C.to_csv('../Project_1_Team_2/general_population_suicide_methods.csv',index=False)
```
